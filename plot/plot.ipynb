```python
import matplotlib.pyplot as plt
 
x1 = ['10', '100', '1000', '10000']
 
y1 = [0.618841333008, 0.786279125977, 0.787717578125, 0.787231494141]
y2 = [0.769331640625, 0.77781862793, 0.747213232422,0.8087015625]
y3 = [0.796088769531, 0.788572387695, 0.793756152344, 0.7371578125]
 
# plt.figure(figsize=(16, 4))
 
plt.title("Latency")
 
plt.plot(x1, y1, label='lock: 10', color='r', marker='o',
         markerfacecolor='red')
plt.plot(x1, y2, label='lock: 100', color='b', marker='*',
         markerfacecolor='blue')
plt.plot(x1, y3, label='lock: 1000', color='g', marker='^',
         markerfacecolor='green')
 
plt.xlabel('Key Range')
 
plt.ylabel('Latency (ms)')
 
# for a, b in zip(x1, y1):
#     plt.text(a, b, b, ha='center', va='bottom', fontsize=20)
 
plt.legend()
plt.savefig("latency.png", dpi=2000)
plt.show()
```
